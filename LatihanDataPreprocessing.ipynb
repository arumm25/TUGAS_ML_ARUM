# Memuat Data

```python
# import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load dataset
df = pd.read_csv('automobileEDA.csv')
```

# Visualisasi Data

```python
# pie chart
fuel_counts = df[['gas', 'diesel']].sum()
labels = ['Gas', 'Diesel']
colors = ['skyblue', 'lightcoral']

plt.figure(figsize=(6, 6))
plt.pie(fuel_counts, labels=labels, autopct='%1.1f%%', colors=colors, startangle=140)
plt.title('Distribusi Jenis Bahan Bakar')
plt.show()
```

```python
# bar chart
plt.figure(figsize=(8, 5))
sns.barplot(x=df['body-style'], y=df['price'], estimator=lambda x: sum(x) / len(x), palette='coolwarm')
plt.xticks(rotation=45)
plt.xlabel('Body Style')
plt.ylabel('Rata-rata Harga')
plt.title('Rata-rata Harga Mobil Berdasarkan Body Style')
plt.show()
```

```python
# scatter plot
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df['engine-size'], y=df['price'], alpha=0.7, color='teal')
plt.xlabel('Engine Size')
plt.ylabel('Price')
plt.title('Scatter Plot: Engine Size vs Price')
plt.show()
```

```python
# heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Heatmap Korelasi Antar Variabel Numerik')
plt.show()
```
