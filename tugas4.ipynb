```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("top100_kdrama.csv")
print("First 5 rows of dataset:")
print(df.head())
```

```python
print("\nDataset Information:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe(include='all'))
```

```python
label_encoder = LabelEncoder()
df['Content Rating'] = label_encoder.fit_transform(df['Content Rating'].astype(str))
```

```python
print("\nMean Number of Episodes:", df['Number of Episode'].mean())
print("Median Rating:", df['Rating'].median())
print("Most Common Genre:", df['Genre'].mode()[0])
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(df['Rating'], bins=10, kde=True, color='blue')
plt.title("Rating Distribution")
plt.show()

sns.boxplot(x=df['Number of Episode'])
plt.title("Episode Count Distribution")
plt.show()

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Feature Correlation")
plt.show()

print("\nScript completed successfully!")
```
